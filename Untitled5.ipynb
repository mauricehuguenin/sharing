### Creating a spatial map with cartopy

```python
save = '/g/data/e14/mv7494/HeatDist_figures/'

# preamble
import xarray as xr # for working with labelled multi-dimensional arrays                                
import cartopy.crs as ccrs # for maps                                                                   
import cartopy.feature as cft
import cartopy.mpl.ticker as cticker
import cartopy.feature as cfeature
import matplotlib.pyplot as plt # for matlab-like plotting                                              
import matplotlib.ticker as ticker                                
import numpy as np # for numerical operations                                                       
import cmocean as cm
```

### Load in some data

```python
%%time
base = '/g/data/e14/mv7494/'
sst_anom = xr.open_dataset(base + 'sst.day.anom.2020-09.nc').anom.isel(time = 26) # select 2020-09-27
sst_anom.plot()
```

### fancy_plot() function for plotting

```python
def fancy_plot(axes_nr):
    '''
    This function creates my specifics for the global plot, i.e.
    coastlines, land colour, gridlines and labels (30lat, 60lon)
    input:
        - axes_nr -> number of subplot, i.e. ax1 for subplot(2,2,1)
    '''
    axes_nr.coastlines(resolution = '50m') # add coast lines
    # ticklabels on left and bottom only
    axes_nr.set_xticks([0,90,180,270,360], crs=ccrs.PlateCarree())   
    axes_nr.set_yticks([-90,-45,0,45,90], crs=ccrs.PlateCarree())
    
    lon_formatter = cticker.LongitudeFormatter();
    lat_formatter = cticker.LatitudeFormatter()
    axes_nr.xaxis.set_major_formatter(lon_formatter) # set format of x-axis
    axes_nr.yaxis.set_major_formatter(lat_formatter) # "       "       "
    axes_nr.add_feature(cfeature.LAND, edgecolor = 'black', facecolor = [.83,.83,.83], zorder = 2)

    axes_nr.xaxis.label.set_visible(False) # no axis labels
    axes_nr.yaxis.label.set_visible(False)

    axes_nr.tick_params(axis='both', which='major', labelsize=16) # fontsize of lon/lat labels
```

### Plotting routine

```python
%%time
# initialise figure
fig = plt.figure(figsize=(9, 9),tight_layout=True,facecolor='w',edgecolor='k')
 
# longitude starting at Cape Agulhas (20E) so that I have Indian Ocean, Pacific Ocean, Atlantic Ocean without breaks
ax1 = plt.subplot(1,1,1, projection=ccrs.PlateCarree(central_longitude=200))
ax1.set_extent([1, 360, -90, 90], crs=ccrs.PlateCarree()) # extent of map

ticks = np.linspace(-4,4,21) # colour bar from -4C to 4C with 21 colours

# the main command here, plotting filled contours of the data
p2 = (sst_anom).plot.contourf(cmp='RdBu_r',levels=ticks,add_colorbar=False,extend='both',transform=ccrs.PlateCarree())


plt.title('NOAA SST Anomalies 27. 09. 2020', fontsize=16) # title

fancy_plot(ax1) # call the function from above that plots labels, coastlines, etc.

    
# add colour bar
# # add colour bar
cb = plt.colorbar(p2, shrink=.35, orientation='vertical'); cb.set_label(label='(C)', size=16)
cb.ax.tick_params(labelsize=16); tick_locator = ticker.MaxNLocator(nbins=5)
cb.locator = tick_locator;cb.update_ticks()

# ---------------------------------------------------------------------------------------------------------------- #
# --- saving as 300 dpi .PNG image in specified folder ----------------------------------------------------------- #
plt.savefig(save + 'noaa_sst_anomalies_2020-09-27', dpi=300, facecolor='w', edgecolor='w', 
            orientation='landscape', papertype=None, format=None, transparent=False,pad_inches=0.1, metadata=None)
# --- end of script ---------------------------------------------------------------------------------------------- #
```
