### Plotting GMSST from the HadISST product over 

HadISST1.1_sst_1870on.pp.gz (~210Mb) downloaded from https://www.metoffice.gov.uk/hadobs/hadisst/

```python
%%time
import xarray as xr
base_Had = '/g/data/e14/mv7494/'                           
HadISST = xr.open_dataset(base_Had + 'HadISST_sst.nc').sst
HadISST = HadISST.where(HadISST != -1e3).mean(['latitude','longitude']).groupby('time.year').mean('time') # replace all fill values with NaN, take global and annual mean
HadISST.plot()
```
